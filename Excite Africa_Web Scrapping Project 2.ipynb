```python
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
```

```python
link2='https://en.wikipedia.org/wiki/List_of_Testudines_families'
response=requests.get(link2)
response.status_code
```

```python
soup=BeautifulSoup(response.text, 'html.parser')
print(soup.prettify())
```

```python
soup.find_all('table')
```

```python
table=soup.find('table', class_='wikitable')
table
```

```python
table_head=table.find_all('tr')[1]
th_=table_head.find_all('th')
col=[t.text.strip() for t in th_]
col
```

```python
content=table.find_all('tr')[2:]
row_ele=[]
for ele in content:
    row_info=ele.find_all('td')
    element=[item.text.strip() for item in row_info]
    row_ele.append(element)
```

```python
data=pd.DataFrame(row_ele,columns=col)
data
```
